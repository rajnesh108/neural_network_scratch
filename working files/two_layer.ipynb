```python
import numpy as np

inputs = np.array([1.0, 2.0, 3.0, 2.5])

weights1 = np.array([
    [0.2, 0.8, -0.5, 1.0],
    [0.5, -0.91, 0.26, -0.5],
    [-0.26, -0.27, 0.17, 0.87]
])
biases1 = np.array([2.0, 3.0, 0.5])

weights2 = np.array([
    [0.1, -0.3, 0.5],
    [-0.2, 0.4, -0.1],
    [0.7, 0.2, -0.6]
])
biases2 = np.array([0.5, -0.5, 1.0])

layer1_dense = np.dot(weights1, inputs) + biases1
layer1_output = np.maximum(0, layer1_dense)

layer2_dense = np.dot(weights2, layer1_output) + biases2

exp_vals = np.exp(layer2_dense - np.max(layer2_dense))
softmax_output = exp_vals / np.sum(exp_vals)

print("Final probabilities:", softmax_output)
```
