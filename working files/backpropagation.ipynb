# Backpropagation

> What is Backpropagation?

       Backpropagation = how a neural network learns

> “How should each weight change to reduce the error?”

        `FORWARD PASS
        Inputs → Layer 1 → ReLU → Layer 2 → Softmax → Prediction`
        
        `BACKWARD PASS
        Error → Layer 2 → Layer 1 → Update weights`

> #### Forward = predict
> #### Backward = learn

        Backpropagation adjusts weights by sending the error backward through the network using gradients.
        
        Too high output? → decrease weights
        Too low output?  → increase weights
        No effect?       → don’t change
        
        Why backprop is powerful?
        
        * Makes deep learning possible
        * Efficient (uses chain rule)
        * Scales to millions of parameters


```python
import numpy as np

# ----------------------------
# 1. Initialization
# ----------------------------
np.random.seed(0)

# Input (1 sample, 4 features)
X = np.array([1.0, 2.0, 3.0, 2.5])

# True label (class 3)
y_true = np.array([0, 0, 1])  # one-hot

# Layer 1 weights (3 neurons, 4 inputs)
W1 = np.array([
    [0.2, 0.8, -0.5, 1.0],
    [0.5, -0.91, 0.26, -0.5],
    [-0.26, -0.27, 0.17, 0.87]
])
b1 = np.array([2.0, 3.0, 0.5])

# Layer 2 weights (3 neurons, 3 inputs)
W2 = np.array([
    [0.1, -0.3, 0.5],
    [-0.2, 0.4, -0.1],
    [0.7, 0.2, -0.6]
])
b2 = np.array([0.5, -0.5, 1.0])

learning_rate = 0.01
epochs = 2

# ----------------------------
# 2. Training loop
# ----------------------------
for epoch in range(epochs):

    # -------- FORWARD PASS --------

    # Layer 1 dense
    z1 = np.dot(W1, X) + b1

    # ReLU activation
    a1 = np.maximum(0, z1)

    # Layer 2 dense
    z2 = np.dot(W2, a1) + b2

    # Softmax activation
    exp_vals = np.exp(z2 - np.max(z2))   # stability
    y_pred = exp_vals / np.sum(exp_vals)

    # print("y_pred: ", y_pred)
    # Loss (Categorical Cross-Entropy)
    loss = -np.log(y_pred[np.argmax(y_true)])
  
    # -------- BACKWARD PASS --------

    # Gradient of loss w.r.t z2 (Softmax + CE)
    dz2 = y_pred - y_true
    # print("dz2: ", dz2)
    
    # Gradients for layer 2
    dW2 = np.outer(dz2, a1)
    # print("dW2: ", dW2)
    db2 = dz2

    # Backprop into layer 1
    da1 = np.dot(W2.T, dz2)

    # ReLU backward
    dz1 = da1 * (z1 > 0)

    # Gradients for layer 1
    dW1 = np.outer(dz1, X)
    # print("dW1: ", dW1)
    db1 = dz1

    # -------- UPDATE WEIGHTS --------

    W2 -= learning_rate * dW2
    b2 -= learning_rate * db2

    W1 -= learning_rate * dW1
    b1 -= learning_rate * db1

    # Print progress
    if epoch % 200 == 0:
        print(f"Epoch {epoch}, Loss: {loss:.4f}")

# ----------------------------
# 3. Final Output
# ----------------------------
print("\nFinal prediction probabilities:")
print(y_pred)
print("Predicted class:", np.argmax(y_pred) + 1)
```
