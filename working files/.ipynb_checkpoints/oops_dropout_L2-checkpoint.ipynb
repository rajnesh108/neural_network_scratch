### What we are going to implement:

1️⃣ Class-based OOP structure
2️⃣ Dropout layer
3️⃣ L2 regularization
4️⃣ 3 hidden layers deep network

Input (4 features)

↓ Dense1 (16 neurons) + ReLU
↓ Dropout

↓ Dense2 (16 neurons) + ReLU
↓ Dropout

↓ Dense3 (16 neurons) + ReLU

↓ Dense4 (3 neurons)
↓ Softmax

Loss: Cross-Entropy + L2 Regularization
Optimizer: Adam

```python
import numpy as np

np.random.seed(0)

# -----------------------------
# ACTIVATION FUNCTIONS
# -----------------------------

def relu(x):
    return np.maximum(0, x)

def relu_backward(dA, Z):
    return dA * (Z > 0)

def softmax(x):
    exp_vals = np.exp(x - np.max(x, axis=1, keepdims=True))
    return exp_vals / np.sum(exp_vals, axis=1, keepdims=True)

# -----------------------------
# DENSE LAYER CLASS
# -----------------------------

class Dense:

    def __init__(self, n_inputs, n_neurons):
        self.W = 0.01 * np.random.randn(n_inputs, n_neurons)
        self.b = np.zeros((1, n_neurons))

        # Adam variables
        self.mW = np.zeros_like(self.W)
        self.vW = np.zeros_like(self.W)
        self.mb = np.zeros_like(self.b)
        self.vb = np.zeros_like(self.b)

    def forward(self, X):
        self.X = X
        self.Z = np.dot(X, self.W) + self.b
        return self.Z

    def backward(self, dZ):

        self.dW = np.dot(self.X.T, dZ)
        self.db = np.sum(dZ, axis=0, keepdims=True)

        return np.dot(dZ, self.W.T)

# -----------------------------
# DROPOUT CLASS
# -----------------------------

class Dropout:

    def __init__(self, rate):
        self.rate = rate

    def forward(self, X):
        self.mask = (np.random.rand(*X.shape) > self.rate) / (1 - self.rate)
        return X * self.mask

    def backward(self, dA):
        return dA * self.mask

# -----------------------------
# LOSS FUNCTION
# -----------------------------

def cross_entropy(y_pred, y_true):
    return -np.mean(np.sum(y_true * np.log(y_pred + 1e-8), axis=1))

# -----------------------------
# ADAM OPTIMIZER
# -----------------------------

class Adam:

    def __init__(self, lr=0.01, beta1=0.9, beta2=0.999, eps=1e-8):
        self.lr = lr
        self.beta1 = beta1
        self.beta2 = beta2
        self.eps = eps
        self.t = 0

    def update(self, layer):

        self.t += 1

        # momentum
        layer.mW = self.beta1*layer.mW + (1-self.beta1)*layer.dW
        layer.mb = self.beta1*layer.mb + (1-self.beta1)*layer.db

        # rms
        layer.vW = self.beta2*layer.vW + (1-self.beta2)*(layer.dW**2)
        layer.vb = self.beta2*layer.vb + (1-self.beta2)*(layer.db**2)

        # bias correction
        mW_corr = layer.mW / (1-self.beta1**self.t)
        vW_corr = layer.vW / (1-self.beta2**self.t)

        mb_corr = layer.mb / (1-self.beta1**self.t)
        vb_corr = layer.vb / (1-self.beta2**self.t)

        # update
        layer.W -= self.lr * mW_corr / (np.sqrt(vW_corr) + self.eps)
        layer.b -= self.lr * mb_corr / (np.sqrt(vb_corr) + self.eps)

# -----------------------------
# DATA
# -----------------------------

X = np.array([
    [1.0,2.0,3.0,2.5],
    [2.0,5.0,-1.0,2.0],
    [-1.5,2.7,3.3,-0.8]
])

y = np.array([
    [0,0,1],
    [1,0,0],
    [0,1,0]
])

# -----------------------------
# MODEL
# -----------------------------

dense1 = Dense(4,16)
drop1 = Dropout(0.2)

dense2 = Dense(16,16)
drop2 = Dropout(0.2)

dense3 = Dense(16,16)

dense4 = Dense(16,3)

optimizer = Adam(lr=0.01)

epochs = 1000
l2_lambda = 1e-4

# -----------------------------
# TRAINING LOOP
# -----------------------------

for epoch in range(epochs):

    # ---- FORWARD ----

    z1 = dense1.forward(X)
    a1 = relu(z1)
    a1 = drop1.forward(a1)

    z2 = dense2.forward(a1)
    a2 = relu(z2)
    a2 = drop2.forward(a2)

    z3 = dense3.forward(a2)
    a3 = relu(z3)

    z4 = dense4.forward(a3)
    y_pred = softmax(z4)

    # ---- LOSS ----
    data_loss = cross_entropy(y_pred, y)

    l2_loss = (
        np.sum(dense1.W**2) +
        np.sum(dense2.W**2) +
        np.sum(dense3.W**2) +
        np.sum(dense4.W**2)
    ) * l2_lambda

    loss = data_loss + l2_loss

    # ---- BACKWARD ----

    dZ4 = (y_pred - y) / X.shape[0]

    dA3 = dense4.backward(dZ4)
    dZ3 = relu_backward(dA3, z3)

    dA2 = dense3.backward(dZ3)
    dA2 = drop2.backward(dA2)
    dZ2 = relu_backward(dA2, z2)

    dA1 = dense2.backward(dZ2)
    dA1 = drop1.backward(dA1)
    dZ1 = relu_backward(dA1, z1)

    dense1.backward(dZ1)

    # L2 gradients
    dense1.dW += l2_lambda * dense1.W
    dense2.dW += l2_lambda * dense2.W
    dense3.dW += l2_lambda * dense3.W
    dense4.dW += l2_lambda * dense4.W

    # ---- UPDATE ----
    optimizer.update(dense1)
    optimizer.update(dense2)
    optimizer.update(dense3)
    optimizer.update(dense4)

    if epoch % 200 == 0:
        print(f"Epoch {epoch} Loss {loss:.4f}")

print("\nFinal Predictions:")
print(np.round(y_pred,3))
```
