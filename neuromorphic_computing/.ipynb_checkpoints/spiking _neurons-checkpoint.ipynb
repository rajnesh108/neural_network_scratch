## 🧠 What is a Spiking Neuron?

A spiking neuron is a mathematical model that imitates how biological neurons work.

Instead of sending continuous values, neurons send spikes (short electrical pulses).

Think of it like:

Input signals → neuron integrates → threshold reached → spike fired

This behavior is inspired by the Action Potential in real biological neurons.

##⚡ How a Spiking Neuron Works

Step-by-step:

1️⃣ Neuron receives input signals
2️⃣ Membrane potential increases
3️⃣ If potential ≥ threshold → neuron fires spike
4️⃣ Potential resets

What happens here:

Voltage slowly increases

When it reaches 0.8

Neuron fires spike

Voltage resets to 0

> Most SNN simulations use the Leaky Integrate-and-Fire Model.

```python
from brian2 import *

start_scope()

tau = 10*ms

eqs = '''
dv/dt = (1 - v)/tau : 1
'''

G = NeuronGroup(1, eqs, threshold='v>0.8', reset='v=0', method='exact')

M = StateMonitor(G, 'v', record=True)

run(100*ms)

plot(M.t/ms, M.v[0])
xlabel('Time (ms)')
ylabel('Membrane potential')
show()
```
