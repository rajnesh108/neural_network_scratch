```python
from brian2 import *
import matplotlib.pyplot as plt

start_scope()

# Create spike times
spike_times = [5, 10, 15, 20, 25]*ms
indices = [0,0,0,0,0]

# Spike generator neuron
G = SpikeGeneratorGroup(1, indices, spike_times)

# Monitor spikes
M = SpikeMonitor(G)

run(30*ms)

# Plot spike train
plt.plot(M.t/ms, M.i, 'k|')
plt.xlabel("Time (ms)")
plt.ylabel("Neuron index")
plt.title("Spike Train Visualization")
plt.show()
```
